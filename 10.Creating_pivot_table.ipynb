```python
import pandas as pd
```

```python
#reading the csv file
df_population_raw = pd.read_csv("population_total.csv")
df_population_raw
```

### 1. Making a Pivot table

pivot(): Returns reshaped Dataframe organinsed by given index/ column values('pivot without aggregation)

```python
#dropping null values
df_population_raw.dropna(inplace=True)
df_population_raw
```

```python
#make a pivod-t table
df_pivot = df_population_raw.pivot(index='year',
                       columns='country',
                       values='population')
df_pivot
```

```python
#selecting some countries and overwrite it
df_pivot = df_pivot[['United States', 'China', 'Rwanda', 'Malta']]
```

```python
#show pivot table
df_pivot
```

## Data Visualisation with pandas


### 1. Lineplot

```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
df_pivot.plot(kind='line')
```

```python
#introduce xlabel, ylabel, title, figsize
df_pivot.plot(kind='line', xlabel='Year', 
              ylabel='Population',
             title='population(1955-2020)',
             figsize=(8,4))

#save plot
plt.savefig('my_project.png')

#show plot
plt.show()
```

```python
#export pivot table to excel
df_pivot.to_excel('pivot_table.xlsx')
```

### 2. Barplot


### 2.1 Single Barplot



```python
#selecting only one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
#transpose dataframe from rows to columns or vice versa
df_pivot_2020.T
```

```python
#overwrite the dataframe
df_pivot_2020 = df_pivot_2020.T
df_pivot_2020
```

```python
#making bar plot
df_pivot_2020.plot(kind='bar')
```

```python
#customise the bar plot, add color, xlabel, ylabel, title
df_pivot_2020.plot(kind='bar', color='orange',
                   xlabel='Country',
                  ylabel='Population',
                  title='Population 2020')
```

```python
#selecting a few years (1980,1990,2000,2010,2020)
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
df_pivot_sample
```

```python
#making grouped barplot
df_pivot_sample.plot(kind='bar', xlabel='Year',
                     ylabel='Population',
                    title='Popluation(1980,1990,2000,2010,2020)')
```
